```python
from random import sample
from collections import Counter


class PLANE:
    def __init__ (self, unique_ID, departed_status, arrived_status):
        self.unique_ID = unique_ID
        self.departed_status = departed_status
        self.arrived_status = arrived_status
        

    def __str__ (self):
        return f"PLANE: {self.unique_ID}, {self.departed_status}, {self.arrived_status}\n"
        
    def __repr__ (self):
        return self.__str__()
    
    def generate_dep_arr(lst_size):
        y = [x for x in range(0,int(lst_size))] + [x for x in range(0,int(lst_size))]
        return sample(y, 2*lst_size - 1)

class ARR_DEP:
    def __init__ (self, num_planes):
        self.num_planes = num_planes
        self.planes = None
        self.y = None
    
    def run(self):
        
        self.y = PLANE.generate_dep_arr(self.num_planes)
        self.planes = [PLANE(i,0,0) for i in range(self.num_planes)]
        
        
        
        for plane_id in self.y:
            plane = self.planes[plane_id]
            #print(self.plane)
            if plane.arrived_status == 0:
                plane.arrived_status += 1
            else:
                plane.departed_status += 1
          
    def find_plane(self):
        for plane in self.planes:
            if plane.departed_status == 0:
                return plane

#planes = ARR_DEP(5)
planes = ARR_DEP(10000)
planes.run()
print(f"Arrival and departure of planes are: {planes.y}")
print(f"The missing plane is: {ARR_DEP.find_plane(planes)}")

#This way is faster
```
