```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_gaussian_quantiles
from sklearn.svm import SVC
```

```python
X,Y = make_gaussian_quantiles(n_samples=1000,n_features=3,n_classes=3)
#print(Y)

A = zip(X,Y)
#print(list(A))
```

```python
clf = SVC()
clf.fit(X,Y)
predicted = clf.predict(X)
#print(predicted)
s = clf.score(X, Y)
print(s)
```

```python
from itertools import product
p = product(range(-3,4), repeat = 3)
k = list(p)
B = np.array(k)
#print(B)
predicted = clf.predict(B)
#print(predicted)
```

```python
fig = plt.figure(num=1, figsize=(4, 4), dpi=200)
ax = fig.add_subplot(projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], marker="o", c=Y, s=25, edgecolor="k")
plt.show()
```
